2. В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью, получены опытные данные: 6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1 Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей, оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.

```python
from statistics import mean
from statistics import pstdev

n = 10
x = [6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1]
m = mean(x)
std = pstdev(x)
t_half_alfa = 2.262
interval = t_half_alfa * std / n**0.5

print(f'Доверительный интервал = {m} ± {interval}')
```

3. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм.
Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр
оказался равным 17.5 мм, а дисперсия известна и равна 4 кв.мм.

```python
d = 4
std = d**0.5
n = 100
# Этапы проверки гипотез:
# 1) формулировка основной и альтернативной гипотез:
# H0: m = m0, H1: m != m0
m0 = 17
m = 17.5
# 2) выбор статистического критерия, на основе которого будет проводиться проверка:
z_alfa = 1.654
z_observ = (m - m0) / std / n**0.5  
# 3) выбор уровня значимости α = 0.05
# 4) определение границ области данной гипотезы:
print(f'Z набл. = {z_observ}')
print(f'Z = {z_half_alfa}')
# 5) подведение итогов и формулировка вывода:
print(f'Z набл < Z => H0 верна')
```

4. Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет: 202, 203, 199, 197, 195, 201, 200, 204, 194, 190. Известно, что их веса распределены нормально. Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%?

```python
x = [202, 203, 199, 197, 195, 201, 200, 204, 194, 190]
std = pstdev(x)
print(f'Оценка стандартного отклонения = {std}')
n = 10
# Этапы проверки гипотез:
# 1) формулировка основной и альтернативной гипотез:
# H0: m = m0, H1: m != m0
m0 = mean(x)
m1 = 200
print(f'Оценка мат. ожидания = {m0}')
# 2) выбор статистического критерия, на основе которого будет проводиться проверка:
t_alfa = 2.821
t_observ = (m1 - m0) / std / n**0.5
# 3) выбор уровня значимости α = 0.01
# 4) определение границ области данной гипотезы:
print(f'P набл. = {t_observ}')
print(f'P = {t_alfa}')
# 5) подведение итогов и формулировка вывода:
print(f'P набл < -P => H0 верна')
```
