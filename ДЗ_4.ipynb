1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.

```python
a = 200
b = 800
M = (a + b)/2
D = (b - a)**2/12
print(f'Среднее значение = {M}')
print(f'Дисперсия = {D}')
```

4. Рост взрослого населения города X имеет нормальное распределение. Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см. Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
а). больше 182 см
б). больше 190 см
в). от 166 см до 190 см
г). от 166 см до 182 см
д). от 158 см до 190 см
е). не выше 150 см или не ниже 190 см
ё). не выше 150 см или не ниже 198 см
ж). ниже 166 см.

```python
from statistics import NormalDist


M = 174
std = 8
height = NormalDist(174, 8)
print(f'a) Рост больше больше 182 см: {(1 - height.cdf(182))*100:.2f}%')
print(f'б) Рост больше больше 190 см: {(1 - height.cdf(190))*100:.2f}%')
print(f'в) Рост от 166 см до 190 см: {(height.cdf(190) - height.cdf(166))*100:.2f}%')
print(f'г) Рост от 166 см до 182 см: {(height.cdf(182) - height.cdf(166))*100:.2f}%')
print(f'д) Рост от 158 см до 190 см: {(height.cdf(190) - height.cdf(158))*100:.2f}%')
print(f'е) Рост не выше 150 см или не ниже 190 см: {(1 - height.cdf(190) + height.cdf(150))*100:.2f}%')
print(f'ё) Рост не выше 150 см или не ниже 198 см: {(height.cdf(198) - height.cdf(150))*100:.2f}%')
print(f'ж) Рост ниже 166 см: {height.cdf(166)*100:.2f}%')
```
